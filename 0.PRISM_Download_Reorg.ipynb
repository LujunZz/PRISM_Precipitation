```python
def ReadBilFile(bil):
    import gdal
    gdal.GetDriverByName('EHdr').Register()
    img = gdal.Open(bil)
    geotransform = img.GetGeoTransform()
    band = img.GetRasterBand(1)
    data = band.ReadAsArray()
    #img.close()
    return data, geotransform
```
